```python
# 중첩 반복문으로 피라미드 만들기
output = ''
for i in range(1, 10):
    for j in range(i):
        output += '*'
    output += '\n'
print(output)
print()

for i in range(1, 10):
    output = ''
    for j in range(i):
        output += '*'
    print(output)
print()

output = ''
for i in range(1, 10):
    output += '*' * i
    output += '\n'
print(output)
print()
```

```python
# 반복문 : while 표현식 : -> 표현식이 True인 동안 실행
# 반복문 탈출 : break, 반복문 계속 : continue
while True:
    for _ in range(10):
        print('.', end='')
        break
    continue # while 문장으로 분기
    break # 실행 안됨
```

```python
# 학생의 이름과 성적을 입력받아 딕셔너리 저장한 후
# 학생의 이름에 'q'가 입력되면 종료
# 학생의 총 인원수와 총점, 평균

sudents = {}
while True:
    student = input("학생과 성적을 입력(ex : 홍길동 90 형식 q는 종료) > ")
    if student == 'q':
        break
    student = student.split()
    if len(student) != 2:
        print("\'학생 성적\' 으로 입력해 주세요")
        continue
    if not student[1].isdecimal():
        print("성적은 숫자로 입력해 주세요")
        continue
    sudents[student[0]] = int(student[1])
```

```python
tot = 0
for key in sudents:
    tot += sudents[key]
print(f'총 인원수 : {len(sudents)}명, 총점 : {tot}, 평균 : {tot/len(sudents):3.2f}')
```

```python
# while 반복문 : 시간을 기반으로
import time

targget_time = time.time() + 5
cnt = 0
while time.time() < targget_time:
    cnt += 1
print(f'5초동안 실행한 반복횟수 : {cnt}')
```

## Chap04-04 문자열, 리스트, 딕셔너리와 관련된 기본함수

```python
numbers = [102, 56, 78, 12, 30]
#numbers에서 가작 작은 값을 출력
min_num = numbers[0]
for num in numbers[1:]:
    if num < min_num:
        min_num = num
print("주어지 리스트에서 가장 작은 값은 : ", min_num)

#==
print(min(numbers))
print(min(34, 12, 53))
```

```python
# 리스트의 합을 구함
tot = 0
for num in numbers:
    tot += num
print(tot)
print(sum(numbers))
```

```python
# names = [], scores = []
# 1. 학생수 입력
# 2. 학생수 만큼 names = [], scores = [] 저장
# 3. 학생 이름으로 성적 출력

names = []
scores = []
while True:
    input_date = input("학생과 성적을 입력(ex : 홍길동 90 형식 q는 종료) > ")
    if input_date == 'q':
        break
    name, score = input_date.split()
    names.append(name)
    scores.append(score)
print(names)
print(scores)
```

```python
s_name = input("검색할 이름 > ")
if s_name in names:
    for idx, value in enumerate(names):
        if value == s_name:
            break
    print(f'{s_name} : {scores[idx]}')
else:
    print("자료없음")
```

```python
# 딕셔너리에 키와 값을 동시에 반환하는 함수 : items()
dict_a = {'key1' : '값1', 'key2' : '값2', 'key3' : '값3'}
for key, value in dict_a.items():
    print(f'{key} : {value}')
```

```python
character = {
    "name" : "기사",
    "leve" : 12,
    "items" : {
        "sword" : "불꽃의 검",
        "armor" : "풀플레이트"
    },
    "skill" :["베기", "세게 베기", "아주 세게 베기"]
}

for key, value in character.items():
    if type(value) is list:
        for item in value:
            print(f'{key} : {item}')
    elif type(value) is dict:
        for val_key in value:
            print(f'{val_key} : {value.get(val_key)}')
    else:
        print (f"{key} : {value}")
```

```python
# 0 부터 19 까지의 수 중 2의 배수의 제곱을 리스트로
array = []
for i in range(0, 20, 2):
    array.append(i*i)
print(array)
print()
# 리스트 내포를 사용하여 변경
# [값 for 반복자 in 반복자료]
array_1 = [i*i for i in range(0, 20, 2)]
print(array_1)

# [값 for 반복자 in 반복자료 if 조건식]
array_2 = [i*i 
           for i in range(0, 20) 
           if i % 2 == 0]
print(array_2)
```
